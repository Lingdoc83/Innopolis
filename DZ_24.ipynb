### Метод TRANSFORM

```python
def get_lower_boundary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return lower_bound

def get_upper_boundary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return upper_bound
```

```python
df_bike['lb'] = df_bike.groupby('Brand')['Ex_Showroom_Price'].transform(get_lower_boundary)
df_bike['ub'] = df_bike.groupby('Brand')['Ex_Showroom_Price'].transform(get_upper_boundary)
```
