### GroupBy

```python
### Группируем таблицу по году выпуску байка и определяем какое количество в продаже
```

```python
df_bike.groupby(['Year'])['Model'].count()
```

### GroupBy + Agg

```python
### Группируем таблицу по бренду с минимальной и максимальной ценой и количеством продаваемых мотоциклов в продаже
```

```python
(
    df_bike
    .groupby('Brand')
    .agg({'Selling_Price': ['min', 'max', 'count']})
)
```

```python
### Таким образом мы узнали, что самые продаваемые бренды Bajaj, Hero и Honda.
```

###Следовательно по своему бюджету лучше всего можно выбрать для покупки байк с 2015 года так как этих мбайков больше всего в подаже брендов Bajaj, Hero или Honda 

```python
df_bike.pivot_table(values = 'KM_Driven', index = 'Brand', columns = 'Owner', aggfunc = 'mean')
```
