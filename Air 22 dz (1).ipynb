### Проверяем выбросы

```python
pd_air_new['late_aircraft_delay'].describe()
```

```python
pd_air_new['late_aircraft_delay'].hist(bins = 100)
```

```python
q1 = pd_air_new['late_aircraft_delay'].quantile(0.25)
q3 = pd_air_new['late_aircraft_delay'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
print(f"Выбросами будут считаться задержка рейсов, у которых значение находится за пределами интервала [{lower_bound}, {upper_bound}]")
```

### Выводим таблицу выбросов

```python
pd_air_new[(pd_air_new['late_aircraft_delay']<=lower_bound)|(pd_air_new['late_aircraft_delay']>=upper_bound)]
```

### Таблица без выбросов

```python
pd_air_new[(pd_air_new['late_aircraft_delay']>=lower_bound)&(pd_air_new['late_aircraft_delay']<=upper_bound)]
```

```python
df_no_outliers = pd_air_new[(pd_air_new['late_aircraft_delay']>=lower_bound)&(pd_air_new['late_aircraft_delay']<=upper_bound)]
```

```python
df_no_outliers['norm_late_aircraft_delay'] = (df_no_outliers['late_aircraft_delay'] - df_no_outliers['late_aircraft_delay'].min()) / (df_no_outliers['late_aircraft_delay'].max() - df_no_outliers['late_aircraft_delay'].min()) 
```
